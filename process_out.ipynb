```python
import pandas as pd

data_df = data["dataset"].to_pandas()
data_ref_df = pd.DataFrame(data["references"])
data_pred_df = pd.DataFrame(data["predictions"])
```

```python
data_ref_df.head()
```

```python
data_pred_df.head()
```

```python
data_df.head()
```

```python
df = pd.merge(pd.merge(data_df, data_pred_df, on="id"), data_ref_df, on="id")
```

```python
df.head()
```

```python
correct = df.apply(lambda x: True if x["prediction_text"] in x["answers_x"]["text"] else False, axis=1)
```

```python
df["correct"] = correct
incorrect_df = df[df["correct"] == False]
```

```python
incorrect_df.shape
```

```python
incorrect_df.sample(500, random_state=42).to_csv("./500_incorrect.csv")
```
